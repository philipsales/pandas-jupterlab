```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
df2
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
df3
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])
df4
```

# **_CONCAT()_**

```python
frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

## **KEYS**

```python
result = pd.concat(frames, keys=['x', 'y', 'z'])
result
```

```python
result.loc['y']
```

## **KEYS as dict**

```python
pieces = {'x': df1, 'y': df2, 'z': df3}
result = pd.concat(pieces)
result
```

```python
result = pd.concat(pieces, keys=['x','z'])
result
```

## **LEVELS**

```python
result.index.levels
```

```python
result = pd.concat(pieces, keys=['x', 'y', 'z'],
                   levels=[['z', 'y', 'x', 'w']],
                    names=['group_key'])
result
```

## **AXIS**

```python
result = pd.concat(frames, axis=1)
result
```

```python
horizontal_join = [df1,df4]
horizon_result = pd.concat(horizontal_join, axis=1, sort=False)
horizon_result
```

## **JOIN**

```python
horizon_result = pd.concat(horizontal_join, axis=1, join='inner')
horizon_result
```

```python
horizon_result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
horizon_result
```

```python
horizon_result = pd.concat([df1, df4], axis=1, join_axes=[df4.index])
horizon_result
```

# **_APPEND()_**

#### **_A useful shortcut to concat() are the append()_**
#### **pd.concat()**
#### **df.append()**

```python
df1
```

```python
df2
```

```python
result = df1.append(df2)
result
```

```python
df4
```

```python
result = df1.append(df4, sort=False)
result
```

```python
result = df1.append([df2, df3, df4], sort=False)
result
```

## **IGNORE_INDEX**

```python
result = pd.concat([df1, df4], ignore_index=True, sort=True)
result
```

```python
s1 = pd.Series(['X0', 'X1', 'X2', 'X3'], name='X')
s1
```

```python
result = pd.concat([df1,s1], axis=1, ignore_index=True)
result
```

## **Mixed Types**

```python
result = pd.concat([df1,s1], axis=0)
result
```

```python
result = pd.concat([df1,s1], axis=1)
result
```
